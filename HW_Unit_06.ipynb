````python
SOWPODS_PATH = "sowpods.txt"

def read_sowpods(fpath: str) -> list[str]:
    """Read the SOWPODS dictionary file and return a set of words."""
    with open(fpath, "r") as f:
        return [line.strip("\n") for line in f.readlines()]


print(f"Read {len(read_sowpods(SOWPODS_PATH))} words from SOWPODS file.")
````

````python
# Q6-2 Grading Tag:

def score_word(word: str) -> int:
    """score_word calculats and return the Scrabble score for that word using the provided scores dictionary."""
    scoring = {"a": 1, "c": 3, "b": 3, "e": 1, "d": 2, "g": 2,
              "f": 4, "i": 1, "h": 4, "k": 5, "j": 8, "m": 3,
              "l": 1, "o": 1, "n": 1, "q": 10, "p": 3, "s": 1,
              "r": 1, "u": 1, "t": 1, "w": 4, "v": 4, "y": 4,
              "x": 8, "z": 10}
    count = {}
    for char in word.lower():
        count[char] = count.get(char, 0) + 1

    score = 0
    characters = list(count.keys())

    i = 0
    while i < len(characters):
        letter = characters[i]
        freq = count[letter]
        if letter in scoring:
            base = scoring[letter]
            letter_total = base * freq
            score += letter_total
        i += 1
    return score
    

def validate_rack(rack: str) -> bool:
    """validate_rack checks if the rack is valid (2-7 characters, only A-Z, a-z, or `*`)"""
    if len(rack) < 2 or len(rack) > 7:
        return False
    
    valid_letters = 'abcdefghijklmnopqrstuvwxyzABCDEFGHIJKLMNOPQRSTUVWXYZ'
    for chars in rack:
        if chars not in valid_letters and chars != '*':
            return False
    return True

def run_scrabble(rack: str, words: list[str]) -> dict[str, int]:
    """run_scrabble finds all valid words and their scores"""
    res = {}
    upper_r = rack.upper()

    for word in words:
        rack_chars, word_chars = sorted(upper_r),sorted(word)
    
        wildcard_c = 0 
        valid = 1
        i = 0
        for char in word_chars:
            if char not in rack_chars:
                 if wildcard_c <upper_r.count('*') and "*" in upper_r:
                    wildcard_c += 1 
                    continue
                 else:
                    valid = 0
                    break
            else:
                rack_chars.remove(char)    

        total_score = 0
        rack_copy = list(upper_r)
        for i in range(len(word)):
            if word[i] in rack_copy:
                rack_copy.remove(word[i])
                total_score += score_word(word[i])
                
        if valid == 1:
            if word in words:
                res[word.upper()] = total_score
        else:
            if "*" in upper_r:
                res[word.upper()] = total_score
    return res


if __name__ == "__main__":
    # The condition above checks if this script is being run directly or being imported. More info:
    # https://docs.python.org/3/library/__main__.html
    # https://realpython.com/if-name-main-python/#in-short-it-allows-you-to-execute-code-when-the-file-runs-as-a-script-but-not-when-its-imported-as-a-module
    words = read_sowpods(SOWPODS_PATH)
    rack = "EXAMPLE"  # feel free to use input() to get the rack from the user.
    print(score_word("EXAMPLE"))
    print(validate_rack(rack))
    print(run_scrabble(rack, words))
````
